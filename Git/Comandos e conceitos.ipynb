# 📚 Git — Guia Prático em Notebook

Este notebook apresenta os principais comandos do **Git** de forma organizada, com explicações, exemplos e boas práticas.

## Objetivos

- Entender o fluxo básico do Git
- Criar e versionar repositórios
- Trabalhar com branches
- Sincronizar com repositórios remotos
- Resolver cenários comuns


## 1. O que é Git?

Git é um sistema de controle de versão distribuído que permite rastrear alterações em arquivos, colaborar com outras pessoas e manter um histórico completo do projeto.

## 2. Configuração inicial

```python
# Configurar identidade
git config --global user.name "Rflsz7"
git config --global user.email "rafaelssm7@gmail.com"

# Ver configurações
git config --list
```

## 3. Criando um repositório

```python
# Criar pasta
mkdir meu-projeto
cd meu-projeto

# Inicializar Git
git init

# Verificar status
git status
```

## 4. Fluxo básico de versionamento

O ciclo mais comum é:

1. Editar arquivos
2. `git status`
3. `git add`
4. `git commit`
5. Repetir

```python
git status
git add .
git commit -m "Primeiro commit"
git log --oneline
```

## 5. Área de Stage

```python
# Adicionar arquivo específico
git add arquivo.py

# Adicionar tudo
git add .

# Remover da área de stage
git restore --staged arquivo.py
```

## 6. Histórico

```python
git log (apertar "q" pra sair)
git log --oneline
git log --graph --oneline --all
```

## 7. Branches

```python
# Listar
git branch

# Criar
git branch feature-login

# Criar e trocar
git switch -c feature-login

# Trocar
git switch main

# Excluir
git branch -d feature-login
```

## 8. Merge

```python
git switch main
git merge feature-login
```

## 9. Repositório remoto (GitHub)

```python
git remote add origin https://github.com/usuario/repositorio.git
git remote -v

git push -u origin main
git pull origin main
git fetch origin
```

## 10. Clonar projeto

```python
git clone https://github.com/usuario/repositorio.git
```

## 11. Desfazendo alterações

```python
# Descartar mudanças locais
git restore arquivo.py

# Restaurar arquivo do último commit
git checkout HEAD -- arquivo.py

# Reverter um commit
git revert <hash>

# Reset (use com cuidado)
git reset --soft HEAD~1
git reset --mixed HEAD~1
git reset --hard HEAD~1
```

## 12. Stash

```python
git stash
git stash list
git stash pop
```

## 13. Tags

```python
git tag v1.0.0
git push origin v1.0.0
git tag
```

## 14. Boas práticas

- Faça commits pequenos e frequentes.
- Escreva mensagens claras.
- Atualize sua branch antes de abrir um merge.
- Evite `git push --force` em branches compartilhadas.
- Use branches para novas funcionalidades e correções.
