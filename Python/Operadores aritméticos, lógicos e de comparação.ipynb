# Operadores em Python

Este notebook explica os principais operadores da linguagem Python:

- Operadores aritméticos
- Operadores de comparação (relacionais)
- Operadores lógicos

Cada seção contém explicações, exemplos práticos, dicas e exercícios.


## 1. Operadores Aritméticos

São utilizados para realizar cálculos matemáticos.


| Operador | Descrição | Exemplo |
|---|---|---|
| `+` | Soma | `5 + 2` |
| `-` | Subtração | `5 - 2` |
| `*` | Multiplicação | `5 * 2` |
| `/` | Divisão | `5 / 2` |
| `//` | Divisão inteira | `5 // 2` |
| `%` | Resto da divisão | `5 % 2` |
| `**` | Potenciação | `5 ** 2` |


```
a = 10
b = 3

print("Soma:", a + b)
print("Subtração:", a - b)
print("Multiplicação:", a * b)
print("Divisão:", a / b)
print("Divisão inteira:", a // b)
print("Resto:", a % b)
print("Potência:", a ** b)
```

### Ordem de precedência

1. `()`
2. `**`
3. `*`, `/`, `//`, `%`
4. `+`, `-`


```
resultado = 5 + 2 * 3
print(resultado)

resultado = (5 + 2) * 3
print(resultado)
```

## 2. Operadores de Comparação

Esses operadores retornam sempre `True` ou `False`.


| Operador | Significado |
|---|---|
| `==` | Igual |
| `!=` | Diferente |
| `>` | Maior |
| `<` | Menor |
| `>=` | Maior ou igual |
| `<=` | Menor ou igual |


```
x = 8
y = 5

print(x == y)
print(x != y)
print(x > y)
print(x < y)
print(x >= 8)
print(y <= 5)
```

## 3. Operadores Lógicos

São utilizados para combinar condições.


| Operador | Descrição |
|---|---|
| `and` | Todas as condições devem ser verdadeiras |
| `or` | Pelo menos uma condição deve ser verdadeira |
| `not` | Inverte o resultado lógico |


```
idade = 20
possui_carteira = True

print(idade >= 18 and possui_carteira)
print(idade < 18 or possui_carteira)
print(not possui_carteira)
```

## 4. Combinando operadores

Na prática, é comum utilizar operadores aritméticos, de comparação e lógicos juntos.


```
nota1 = 8
nota2 = 6

media = (nota1 + nota2) / 2

if media >= 7 and media <= 10:
    print("Aprovado")
else:
    print("Reprovado")
```

## 5. Boas práticas

- Utilize parênteses para deixar expressões complexas mais legíveis.
- Evite condições muito longas.
- Dê nomes claros às variáveis booleanas (`ativo`, `logado`, `possui_permissao`).
- Revise sempre a diferença entre `=` (atribuição) e `==` (comparação).


```
usuario_logado = True
eh_admin = False

if usuario_logado and not eh_admin:
    print("Usuário comum autenticado")
```
