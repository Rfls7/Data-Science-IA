# Dicionários em Python

Este notebook apresenta **dicionários (`dict`) em Python** em detalhes, com exemplos práticos, boas práticas e exercícios.

## 1. O que é um dicionário?

Um dicionário é uma estrutura de dados que armazena pares **chave: valor**.

Características:
- Mutável
- Não permite chaves duplicadas
- Acesso rápido aos valores pelas chaves
- Criado com `{}` ou `dict()`


```python
pessoa = {
    "nome": "Ana",
    "idade": 25,
    "cidade": "Brasília"
}

print(pessoa)
```

## 2. Acessando valores

Acesse um valor utilizando sua chave.

```python
print(pessoa["nome"])
print(pessoa.get("idade"))
print(pessoa.get("telefone"))   # retorna None
```

## 3. Adicionando e alterando

Novas chaves podem ser adicionadas e valores existentes podem ser alterados.

```python
pessoa["profissao"] = "Engenheira"
pessoa["idade"] = 26
print(pessoa)
```

## 4. Removendo elementos

Existem várias formas de remover itens.

```python
dados = {"a":1,"b":2,"c":3}

dados.pop("b")
del dados["a"]
print(dados)
```

## 5. Percorrendo dicionários

Pode-se percorrer chaves, valores ou ambos.

```python
for chave in pessoa:
    print(chave)

for valor in pessoa.values():
    print(valor)

for chave, valor in pessoa.items():
    print(chave, "->", valor)
```

## 6. Métodos úteis

Alguns métodos importantes:
- keys()
- values()
- items()
- update()
- pop()
- clear()
- copy()


```python
d1 = {"x":1}
d2 = {"y":2}

d1.update(d2)
print(d1)
```

## 7. Dicionários aninhados

Um dicionário pode conter outros dicionários.

```python
aluno = {
    "nome":"Carlos",
    "notas":{
        "Python":9,
        "Banco":8
    }
}

print(aluno["notas"]["Python"])
```

## 8. Verificando existência

Use o operador `in`.

```python
if "nome" in pessoa:
    print("Existe a chave nome")
```

## 9. Compreensão de dicionários

É possível criar dicionários de forma compacta.

```python
quadrados = {x: x**2 for x in range(6)}
print(quadrados)
```

## 10. Boas práticas

- Use chaves significativas.
- Evite alterar um dicionário enquanto o percorre.
- Prefira `.get()` quando a chave pode não existir.
- Utilize `items()` quando precisar da chave e do valor.


```python
estoque = {"maçã":10,"banana":5}

for produto, quantidade in estoque.items():
    print(f"{produto}: {quantidade}")
```
