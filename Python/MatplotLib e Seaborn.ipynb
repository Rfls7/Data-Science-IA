## 1. O que são Matplotlib e Seaborn?

O **Matplotlib** é a principal biblioteca de visualização de dados em Python.

O **Seaborn** é construído sobre o Matplotlib e oferece gráficos estatísticos com uma aparência mais moderna e recursos adicionais.


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

print("Matplotlib e Seaborn carregados com sucesso!")
```

## 2. Gráfico de Linha

Ideal para visualizar tendências ao longo do tempo.

```python
x = [1,2,3,4,5]
y = [2,4,3,6,5]

plt.plot(x, y, marker='o')
plt.title("Gráfico de Linha")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

## 3. Gráfico de Barras

Utilizado para comparar categorias.

```python
categorias = ["A","B","C","D"]
valores = [15, 30, 20, 40]

plt.bar(categorias, valores)
plt.title("Gráfico de Barras")
plt.show()
```

## 4. Histograma

Mostra a distribuição de uma variável numérica.

```python
import numpy as np

dados = np.random.normal(50, 10, 1000)

plt.hist(dados, bins=20)
plt.title("Histograma")
plt.show()
```

## 5. Gráfico de Dispersão

Exibe a relação entre duas variáveis.

```python
x = np.random.rand(100)
y = x * 2 + np.random.rand(100)

plt.scatter(x, y)
plt.title("Dispersão")
plt.show()
```

## 6. Personalizando gráficos

É possível alterar títulos, rótulos, legendas, cores e estilos.

```python
plt.plot([1,2,3],[3,2,5], label="Série 1")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.title("Exemplo")
plt.legend()
plt.grid(True)
plt.show()
```

## 7. Introdução ao Seaborn

O Seaborn funciona muito bem com DataFrames do Pandas.

```python
df = pd.DataFrame({
    "Idade":[20,22,25,30,35],
    "Salario":[2000,2500,3000,4200,5000]
})

sns.scatterplot(data=df, x="Idade", y="Salario")
plt.show()
```

## 8. Gráfico de Barras com Seaborn

Utilizado para comparar categorias com agregações estatísticas.

```python
vendas = pd.DataFrame({
    "Produto":["A","A","B","B","C"],
    "Valor":[10,15,20,25,30]
})

sns.barplot(data=vendas, x="Produto", y="Valor")
plt.show()
```

## 9. Boxplot

Mostra mediana, quartis e possíveis outliers.

```python
sns.boxplot(data=df, y="Salario")
plt.show()
```

## 10. Heatmap

Muito utilizado para visualizar matrizes e correlações.

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="Blues")
plt.show()
```

## 11. Pairplot

Permite comparar automaticamente todas as variáveis numéricas.

```python
iris = sns.load_dataset("iris")

sns.pairplot(iris, hue="species")
plt.show()
```

## 12. Boas práticas

- Escolha o gráfico adequado ao tipo de dado.
- Evite excesso de informações.
- Utilize títulos e rótulos claros.
- Use cores para destacar informações importantes, mas com moderação.
- Sempre identifique unidades quando necessário.


```python
sns.set_theme(style="whitegrid")

plt.plot([1,2,3],[2,5,3])
plt.title("Exemplo com tema Seaborn")
plt.show()
```
