## 1. O que é o NumPy?

NumPy é a principal biblioteca para computação numérica em Python. Seu objeto principal é o `ndarray`, que é mais eficiente que listas para operações matemáticas.

```python
import numpy as np

print(np.__version__)
```

## 2. Criando arrays

Arrays podem ser criados a partir de listas ou funções do NumPy.

```python
a = np.array([1,2,3,4])
b = np.zeros(5)
c = np.ones((2,3))
d = np.arange(0,10,2)

print(a)
print(b)
print(c)
print(d)
```

## 3. Propriedades

Descubra dimensões, formato e tipo dos dados.

```python
matriz = np.array([[1,2,3],[4,5,6]])

print(matriz.shape)
print(matriz.ndim)
print(matriz.size)
print(matriz.dtype)
```

## 4. Indexação e fatiamento

Acesse elementos, linhas, colunas e subconjuntos.

```python
arr = np.array([[10,20,30],[40,50,60]])

print(arr[0,1])
print(arr[:,0])
print(arr[1,:])
print(arr[:,1:])
```

## 5. Operações matemáticas

As operações são vetorizadas.

```python
x = np.array([1,2,3])
y = np.array([4,5,6])

print(x+y)
print(x-y)
print(x*y)
print(x/y)
print(x**2)
```

## 6. Broadcasting

Permite operações entre arrays de tamanhos compatíveis.

```python
mat = np.array([[1,2,3],[4,5,6]])
print(mat + 10)
```

## 7. Funções estatísticas

Calcule estatísticas rapidamente.

```python
dados = np.array([10,20,30,40,50])

print(np.mean(dados))
print(np.median(dados))
print(np.std(dados))
print(np.min(dados))
print(np.max(dados))
print(np.sum(dados))
```

## 8. Remodelando arrays

Use `reshape()` para alterar o formato.

```python
v = np.arange(12)
m = v.reshape(3,4)

print(v)
print(m)
```

## 9. Álgebra linear

NumPy possui diversas funções para álgebra linear.

```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

print(A @ B)
print(np.linalg.det(A))
print(np.linalg.inv(A))
```

## 10. Números aleatórios

Gere amostras aleatórias.

```python
np.random.seed(42)

print(np.random.rand(5))
print(np.random.randint(1,11,size=5))
print(np.random.normal(0,1,5))
```

## 11. Boas práticas

- Prefira operações vetorizadas.
- Evite loops quando possível.
- Utilize arrays homogêneos.
- Aproveite funções do módulo `numpy.linalg` e `numpy.random`.


```python
dados = np.arange(1,6)
print(dados * 2)
print(np.sqrt(dados))
```
