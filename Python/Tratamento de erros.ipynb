# Tratamento de Erros e Exceções em Python

Este notebook apresenta os principais conceitos sobre **tratamento de erros (exceptions)** em Python. Você aprenderá a identificar erros, utilizar `try`, `except`, `else`, `finally`, criar exceções personalizadas e aplicar boas práticas.


## 1. O que são erros e exceções?

Durante a execução de um programa podem ocorrer situações inesperadas, como divisão por zero ou conversão inválida de dados.

- **Erro de sintaxe (SyntaxError):** impede o programa de iniciar.
- **Exceção (Exception):** ocorre durante a execução do programa e pode ser tratada.


```python
# Exemplo de exceção
numero = 10
divisor = 0

print(numero / divisor)  # ZeroDivisionError
```

## 2. Bloco try e except

O bloco `try` contém o código que pode gerar uma exceção. O bloco `except` captura e trata o erro.

```python
try:
    numero = int(input("Digite um número: "))
    print(100 / numero)
except:
    print("Ocorreu um erro.")
```

## 3. Capturando exceções específicas

É recomendado capturar apenas as exceções esperadas.

```python
try:
    valor = int(input("Digite um número inteiro: "))
    print(10 / valor)

except ValueError:
    print("Entrada inválida.")

except ZeroDivisionError:
    print("Não é possível dividir por zero.")
```

## 4. Obtendo informações do erro

É possível acessar a mensagem da exceção usando `as`.

```python
try:
    arquivo = open("arquivo_inexistente.txt")
except FileNotFoundError as erro:
    print("Erro encontrado:")
    print(erro)
```

## 5. Bloco else

O bloco `else` é executado quando nenhuma exceção ocorre.

```python
try:
    idade = int(input("Idade: "))
except ValueError:
    print("Valor inválido.")
else:
    print(f"Idade informada: {idade}")
```

## 6. Bloco finally

O bloco `finally` é executado sempre, independentemente de ocorrer erro ou não.

```python
try:
    print("Abrindo recurso...")
except:
    pass
finally:
    print("Fechando recurso.")
```

## 7. Levantando exceções com raise

Use `raise` para gerar exceções manualmente quando uma condição inválida for encontrada.

```python
idade = -5

if idade < 0:
    raise ValueError("A idade não pode ser negativa.")
```

## 8. Criando exceções personalizadas

Você pode criar suas próprias classes de exceção herdando de `Exception`.

```python
class SaldoInsuficienteError(Exception):
    pass

saldo = 100
saque = 150

if saque > saldo:
    raise SaldoInsuficienteError("Saldo insuficiente para realizar o saque.")
```

## 9. Boas práticas

- Capture exceções específicas.
- Evite usar `except:` sem necessidade.
- Não esconda erros importantes.
- Utilize `finally` para liberar recursos.
- Escreva mensagens de erro claras.


```python
def dividir(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Divisão por zero não permitida."

print(dividir(10, 2))
print(dividir(10, 0))
```
