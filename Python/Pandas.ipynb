## 1. O que é o Pandas?

O Pandas é uma biblioteca para análise e manipulação de dados estruturados.

Seus principais objetos são:

- **Series**: estrutura unidimensional.
- **DataFrame**: tabela bidimensional semelhante a uma planilha.

## 3. Criando uma Series

Uma Series é semelhante a uma coluna de uma planilha.

````python
idades = pd.Series([18, 22, 35, 40])
print(idades)
````

## 4. Criando um DataFrame

Um DataFrame organiza dados em linhas e colunas.

````python
dados = {
    "Nome": ["Ana", "Carlos", "Maria"],
    "Idade": [25, 32, 29],
    "Cidade": ["Brasília", "São Paulo", "Recife"]
}

df = pd.DataFrame(dados)
print(df)
````

## 6. Explorando os dados

Métodos úteis para conhecer um conjunto de dados.

````python
print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
````

## 8. Filtrando dados

Filtros utilizam expressões booleanas.

````python
maiores = df[df["Idade"] >= 30]
print(maiores)
````

## 9. Adicionando e removendo colunas

É possível criar ou excluir colunas facilmente.

````python
df["Maior_de_Idade"] = df["Idade"] >= 18

print(df)

del df["Maior_de_Idade"]
````

## 10. Valores ausentes

O Pandas oferece funções para identificar e tratar valores nulos.

````python
import numpy as np

df2 = pd.DataFrame({
    "A":[1,2,np.nan],
    "B":[5,np.nan,8]
})

print(df2.isnull())
print(df2.fillna(0))
````

## 11. Ordenação

Ordene linhas por uma ou mais colunas.

````python
print(df.sort_values("Idade"))
````

## 12. Agrupamento

Use `groupby()` para resumir dados.

````python
vendas = pd.DataFrame({
    "Cidade":["A","A","B","B"],
    "Valor":[100,200,150,300]
})

print(vendas.groupby("Cidade")["Valor"].sum())
````

## 13. Aplicando funções

`apply()` executa funções em colunas ou linhas.

````python
df["Idade em meses"] = df["Idade"].apply(lambda x: x * 12)
print(df)
````
