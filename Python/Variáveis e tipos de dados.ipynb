# Variáveis e Tipos de Dados em Python

Este notebook apresenta os conceitos fundamentais de **variáveis** e **tipos de dados** em Python, com explicações detalhadas, exemplos práticos e exercícios.


## 1. O que são variáveis?

Uma variável é um nome que referencia um valor armazenado na memória.

Em Python, não é necessário declarar o tipo da variável antes de utilizá-la. O tipo é definido automaticamente de acordo com o valor atribuído.


```
nome = "Maria"
idade = 22
altura = 1.68

print(nome)
print(idade)
print(altura)
```

## 2. Regras para nomes de variáveis

Boas práticas:
- Devem começar com uma letra ou `_`.
- Não podem começar com números.
- Não podem conter espaços.
- Diferenciam letras maiúsculas e minúsculas.
- Evite palavras reservadas da linguagem.


```
nome_completo = "João Silva"
idade2 = 30

print(nome_completo)
print(idade2)
```

## 3. Tipos de dados básicos

Os principais tipos nativos são:

- `int` → números inteiros
- `float` → números com casas decimais
- `str` → textos
- `bool` → valores lógicos (`True` ou `False`)


```
inteiro = 10
decimal = 3.14
texto = "Python"
ativo = True

print(type(inteiro))
print(type(decimal))
print(type(texto))
print(type(ativo))
```

## 4. Conversão de tipos (Casting)

É possível converter valores entre diferentes tipos usando funções como `int()`, `float()`, `str()` e `bool()`.


```
numero = "25"

print(int(numero) + 5)

valor = 8
print(float(valor))

idade = 20
print(str(idade))
```

## 5. Entrada de dados

A função `input()` lê dados digitados pelo usuário. Ela sempre retorna uma string.


```
nome = input("Digite seu nome: ")
idade = int(input("Digite sua idade: "))

print(f"Olá, {nome}!")
print(f"Você tem {idade} anos.")
```

## 6. Atribuição múltipla

Python permite atribuir vários valores em uma única linha.


```
x, y, z = 10, 20, 30

print(x)
print(y)
print(z)
```

## 7. Constantes (convenção)

Python não possui constantes reais. Por convenção, utilizamos letras maiúsculas para indicar que um valor não deve ser alterado.


```
PI = 3.14159
VELOCIDADE_DA_LUZ = 299792458

print(PI)
```

## 8. Operador `type()`

A função `type()` retorna o tipo de um objeto.


```
valor = 15.5

print(type(valor))
```

## 9. Boas práticas

- Utilize nomes descritivos.
- Evite abreviações desnecessárias.
- Utilize `snake_case`.
- Inicialize variáveis antes de utilizá-las.
- Escolha tipos adequados para cada informação.


```
quantidade_alunos = 35
media_final = 8.7
curso = "Python"

print(curso)
```
