## 1. O que é POO?

A Programação Orientada a Objetos organiza o código em **objetos**, que representam entidades do mundo real. Cada objeto possui **atributos** (características) e **métodos** (ações).

## 2. Criando uma classe

```python
# Define uma classe chamada Pessoa.
# Neste momento ela não possui atributos nem métodos.
# A instrução 'pass' é usada apenas para indicar que o bloco da classe está vazio.
class Pessoa:
    pass

# Cria um objeto (instância) da classe Pessoa.
# A variável 'p1' passa a referenciar esse novo objeto.
p1 = Pessoa()

# Exibe o tipo do objeto criado.
# O resultado será: <class '__main__.Pessoa'>
print(type(p1))
```

## 3. Atributos e métodos

```python
# Define a classe Pessoa.
class Pessoa:

    # Define um método chamado 'apresentar'.
    # O parâmetro 'self' representa o próprio objeto que está chamando o método.
    def apresentar(self):
        print("Olá! Eu sou uma pessoa.")

# Cria um objeto (instância) da classe Pessoa.
p = Pessoa()

# Chama o método 'apresentar' do objeto 'p'.
# Como o método pertence ao objeto, utilizamos a notação objeto.metodo().
p.apresentar()
```

## 4. O construtor (`__init__`)

```python
# Define a classe Pessoa.
class Pessoa:

    # O método especial __init__ é chamado automaticamente
    # sempre que um novo objeto é criado.
    # Ele é utilizado para inicializar os atributos do objeto.
    def __init__(self, nome, idade):
        self.nome = nome      # Armazena o nome do objeto
        self.idade = idade    # Armazena a idade do objeto

    # Método responsável por apresentar as informações da pessoa.
    def apresentar(self):
        print(f"Meu nome é {self.nome} e tenho {self.idade} anos.")

# Cria um objeto da classe Pessoa.
# Os valores "Ana" e 28 são enviados para o construtor (__init__).
p = Pessoa("Ana", 28)

# Chama o método apresentar() para exibir os dados do objeto.
p.apresentar()
```

## 5. Atributos de instância

```python
class Produto:
    def __init__(self, nome, preco):
        self.nome = nome
        self.preco = preco

produto = Produto("Notebook", 3500)
print(produto.nome)
print(produto.preco)
```

## 6. Atributos de classe

```python
class Conta:
    banco = "Banco Python"

    def __init__(self, titular):
        self.titular = titular

c = Conta("Carlos")
print(c.banco)
print(Conta.banco)
```

## 7. Encapsulamento

```python
# Define a classe Conta.
class Conta:

    # Construtor da classe.
    # Recebe o saldo inicial da conta.
    def __init__(self, saldo):

        # O atributo __saldo é privado.
        # O uso de dois underscores (__) indica que ele não deve ser acessado
        # diretamente fora da classe.
        self.__saldo = saldo

    # Método para depositar um valor na conta.
    def depositar(self, valor):

        # Soma o valor informado ao saldo atual.
        self.__saldo += valor

    # Método para consultar o saldo da conta.
    def consultar_saldo(self):

        # Retorna o saldo armazenado.
        return self.__saldo

# Cria uma conta com saldo inicial de R$ 100.
c = Conta(100)

# Deposita R$ 50 na conta.
c.depositar(50)

# Exibe o saldo atualizado.
print(c.consultar_saldo())
```

## 8. Herança

```python
# Classe base (ou classe pai).
# Ela representa um animal de forma genérica.
class Animal:

    # Método que representa o som emitido por um animal.
    def emitir_som(self):
        print("Au Au!")


# Classe Cachorro.
# Ela herda todas as características da classe Animal.
class Cachorro(Animal):
    pass


# Cria um objeto da classe Cachorro.
dog = Cachorro()

# Chama o método emitir_som().
# A classe cachorro não possui esse método, mas como ela herda da classe Animal,
dog.emitir_som()
```

## 9. Polimorfismo

```python
# Define a classe Gato.
class Gato:

    # Método responsável por emitir o som do gato.
    def emitir_som(self):
        print("Miau")


# Define a classe Vaca.
class Vaca:

    # Método responsável por emitir o som da vaca.
    def emitir_som(self):
        print("Muu")


# Cria uma lista contendo um objeto da classe Gato
# e um objeto da classe Vaca.
for animal in [Gato(), Vaca()]:

    # Chama o método emitir_som() para cada objeto.
    # O Python executa o método correspondente à classe do objeto.
    animal.emitir_som()
```

## 10. Métodos especiais

```python
# Define a classe Livro.
class Livro:

    # Construtor da classe.
    # Recebe o título do livro e o armazena no atributo 'titulo'.
    def __init__(self, titulo):
        self.titulo = titulo

    # Método especial __str__().
    # Ele define como o objeto será representado em formato de texto.
    # Esse método é chamado automaticamente quando utilizamos print().
    def __str__(self):
        return f"Livro: {self.titulo}"

# Cria um objeto da classe Livro e o imprime na tela.
print(Livro("Python para Todos"))
```

## 11. Propriedades (`@property`)

```python
# Define a classe Pessoa.
class Pessoa:

    # Construtor da classe.
    # Recebe a idade e a armazena no atributo _idade.
    def __init__(self, idade):
        self._idade = idade

    # O decorador @property transforma o método 'idade'
    # em um atributo de leitura.
    # Assim, podemos acessar a idade sem usar parênteses.
    @property
    def idade(self):
        return self._idade

# Cria um objeto da classe Pessoa com idade igual a 30.
p = Pessoa(30)

# Acessa a propriedade 'idade' e imprime seu valor.
print(p.idade)
```

## 12. Métodos de classe e estáticos

```python
# Define a classe Calculadora.
class Calculadora:

    # O decorador @staticmethod indica que este método
    # não depende de um objeto nem da própria classe.
    # Ele funciona como uma função comum, mas organizada dentro da classe.
    @staticmethod
    def somar(a, b):
        return a + b

    # O decorador @classmethod indica que este método
    # recebe a própria classe como primeiro parâmetro (cls).
    @classmethod
    def descricao(cls):
        return f"Classe: {cls.__name__}"

# Chama o método estático sem criar um objeto.
print(Calculadora.somar(2, 3))

# Chama o método de classe.
print(Calculadora.descricao())
```

## 13. Resumo

| Conceito | Descrição |
|---|---|
| Classe | Modelo para criar objetos |
| Objeto | Instância de uma classe |
| Atributo | Característica do objeto |
| Método | Ação do objeto |
| `__init__` | Construtor |
| Herança | Reutilização entre classes |
| Polimorfismo | Mesmo método, comportamentos diferentes |
| Encapsulamento | Proteção dos dados |
| `@property` | Acesso controlado a atributos |


## 14. Boas práticas

- Dê nomes claros às classes (PascalCase).
- Prefira métodos curtos e com uma responsabilidade.
- Utilize encapsulamento quando necessário.
- Evite duplicação de código com herança ou composição.
- Documente classes e métodos importantes.

2. Crie uma classe `Retangulo` com um método que calcule a área.

```python

class Retangulo:
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def area(self):
        return self.base * self.altura

# Exemplo de uso da classe Retangulo
r = Retangulo(5, 3)
print(r.area())  # Saída: 15
```
