# Variáveis Locais e Globais em Python

Este notebook explica detalhadamente o conceito de **escopo de variáveis** em Python, abordando variáveis locais, globais, a palavra-chave `global`, a palavra-chave `nonlocal`, boas práticas e exercícios.


## 1. O que é escopo?

O **escopo** define onde uma variável pode ser acessada dentro de um programa.

Os principais tipos são:

- **Escopo global**: variável criada fora de funções.
- **Escopo local**: variável criada dentro de uma função.


```
mensagem = "Olá!"

def exibir():
    print(mensagem)

exibir()
print(mensagem)
```

## 2. Variáveis globais

Uma variável global é criada fora de qualquer função e pode ser lida dentro das funções.

```
contador = 10

def mostrar():
    print(contador)

mostrar()
print(contador)
```

## 3. Variáveis locais

Variáveis locais existem apenas dentro da função onde foram criadas.

```
def saudacao():
    nome = "Ana"
    print(nome)

saudacao()

# print(nome)  # Gera NameError
```

## 4. Mesmo nome em escopos diferentes

Uma variável local pode ter o mesmo nome de uma variável global. A local tem prioridade dentro da função.

```
valor = 100

def teste():
    valor = 50
    print("Dentro da função:", valor)

teste()
print("Fora da função:", valor)
```

## 5. A palavra-chave global

Para alterar uma variável global dentro de uma função, utilize `global`.

```
contador = 0

def incrementar():
    global contador
    contador += 1

incrementar()
incrementar()

print(contador)
```

## 6. A palavra-chave nonlocal

`nonlocal` é utilizada em funções aninhadas para modificar variáveis da função externa.

```
def externa():
    contador = 0

    def interna():
        nonlocal contador
        contador += 1
        print(contador)

    interna()
    interna()

externa()
```

## 7. Escopo LEGB

Python procura variáveis seguindo a regra **LEGB**:

1. Local
2. Enclosing (função externa)
3. Global
4. Built-in


```
x = "Global"

def externa():
    x = "Externa"

    def interna():
        x = "Local"
        print(x)

    interna()

externa()
```

## 8. Boas práticas

- Evite modificar variáveis globais sempre que possível.
- Prefira passar valores como parâmetros.
- Utilize variáveis locais para tornar funções independentes.
- Use `global` apenas quando realmente necessário.


```
def soma(a, b):
    resultado = a + b
    return resultado

print(soma(3, 5))
```
