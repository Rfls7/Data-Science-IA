### Início


```python

# Aqui eu crio uma classe, ou seja, um modelo para criar objetos que representem pessoas com nome e idade.

# Declaração da classe
class Ganhafoto:

    #  Método Construtor para construir os atributos de instância da classe Ganhafoto
    def __init__(self): 
        # Atributos de Instância
        self.nome = ""
        self.idade = 0

    # Métodos de Instância
    def aniversario(self):
        self.idade += 1

    def mensagem(self):
        return f"Olá, meu nome é {self.nome} e tenho {self.idade} anos."
    
# Declaração de um objeto da classe Ganhafoto
g1 = Ganhafoto()

# Atribuição de valores aos atributos do objeto g1
g1.nome = "João"
g1.idade = 25

g1.aniversario()  # Chamando o método aniversario() do objeto g1 que faz com que a idade aumente em 1

# Chamando o método mensagem() do objeto g1
print(g1.mensagem()) 
```

### Melhoria

```python

# Aqui eu crio uma classe, ou seja, um modelo para criar objetos que representem pessoas com nome e idade.

# Declaração da classe
class Ganhafoto:

    """ 
Aqui eu posso escrever como funciona a classe Ganhafoto, o que ela faz, quais são seus atributos e métodos, etc. Inclusive, posso escrever exemplos de como usar a classe.    
    """

    #  Método Construtor para construir os atributos de instância da classe Ganhafoto
    def __init__(self, nome = "", idade = 0): 
        # Atributos de Instância
        self.nome = nome
        self.idade = idade

    # Métodos de Instância
    def aniversario(self):
        self.idade += 1


    # Método especial __str__() que retorna uma string representando o objeto # Dunder method
    def __str__(self):
        return f"Olá, meu nome é {self.nome} e tenho {self.idade} anos."
    
# Declaração de um objeto da classe Ganhafoto
g1 = Ganhafoto(nome="João", idade=25)

g1.aniversario()  # Chamando o método aniversario() do objeto g1 que faz com que a idade aumente em 1

# Chamando o método __str__() do objeto g1
print(g1) 

print(g1.__getstate__())  # Chamando o método __getstate__() do objeto g1 que retorna um dicionário com os atributos do objeto

# Chamando o atributo __doc__ da classe Ganhafoto para exibir a documentação da classe
print(g1.__doc__)
```

### Conta Bancária

```python
class ContaBancaria:

    def __init__(self, id, nome, saldo=0):
        self.id = id
        self.titular = nome
        self.saldo = saldo

        print(
            f"Conta {self.id} criada com sucesso. Saldo atual de R${self.saldo:,.2f}"
        )

    def __str__(self):
        return (
            f"A conta {self.id} de {self.titular} tem R${self.saldo:,.2f} de saldo."
        )

    def depositar(self, valor):
        self.saldo += valor
        print(
            f"Depósito de R${valor:,.2f} autorizado na conta {self.id}"
        )

    def sacar(self, valor):
        if valor > self.saldo:
            print(
                f"Saque NEGADO de R${valor:,.2f} na conta {self.id}: SALDO INSUFICIENTE"
            )
        else:
            self.saldo -= valor
            print(
                f"Saque de R${valor:,.2f} autorizado na conta {self.id}"
            )



c1 = ContaBancaria(112, "Gustavo", 3000)
c1.depositar(500)
c1.sacar(2_000_000)
c1.sacar(200)
print(c1)
```
