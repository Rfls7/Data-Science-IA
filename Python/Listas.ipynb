# Listas em Python

Este notebook explica **listas (`list`) em Python** de forma detalhada, com teoria, exemplos, boas práticas e exercícios.

## 1. O que são listas?

Uma lista é uma estrutura de dados que armazena uma sequência ordenada de elementos.

Características:
- Mutável
- Aceita elementos de diferentes tipos
- Permite elementos repetidos
- Criada utilizando colchetes `[]`


```python
numeros = [10, 20, 30, 40]
nomes = ["Ana", "Carlos", "Maria"]
mista = [1, "Python", True, 3.14]

print(numeros)
print(nomes)
print(mista)
```

## 2. Acessando elementos

Os índices começam em 0. Também é possível utilizar índices negativos.

```python
frutas = ["maçã", "banana", "uva", "laranja"]

print(frutas[0])
print(frutas[2])
print(frutas[-1])
```

## 3. Fatiamento (Slicing)

É possível obter partes da lista usando a sintaxe `inicio:fim:passo`.

```python
lista = [0,1,2,3,4,5,6,7,8]

print(lista[2:6])
print(lista[:4])
print(lista[5:])
print(lista[::2])
```

## 4. Alterando elementos

Como listas são mutáveis, seus elementos podem ser alterados.

```python
cores = ["azul", "verde", "vermelho"]

cores[1] = "amarelo"

print(cores)
```

## 5. Adicionando elementos

Métodos principais: `append()`, `insert()` e `extend()`.

```python
numeros = [1,2,3]

numeros.append(4)
numeros.insert(0, 0)
numeros.extend([5,6])

print(numeros)
```

## 6. Removendo elementos

Métodos comuns: `remove()`, `pop()`, `clear()` e `del`.

```python
dados = [10,20,30,40,50]

dados.remove(20)
dados.pop()
del dados[0]

print(dados)
```

## 7. Percorrendo listas

Use laços `for` ou `enumerate()`.

```python
nomes = ["Ana", "Carlos", "Maria"]

for nome in nomes:
    print(nome)

for indice, nome in enumerate(nomes):
    print(indice, nome)
```

## 8. Métodos importantes

- sort()
- reverse()
- count()
- index()
- copy()


```python
valores = [5,2,8,1,2]

print(valores.count(2))
print(valores.index(8))

valores.sort()
print(valores)

valores.reverse()
print(valores)
```

## 9. List Comprehension

Forma compacta de criar listas.

```python
quadrados = [x**2 for x in range(10)]

pares = [x for x in range(20) if x % 2 == 0]

print(quadrados)
print(pares)
```

## 10. Listas aninhadas

Uma lista pode conter outras listas.

```python
matriz = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

print(matriz[1][2])
```

## 11. Boas práticas

- Prefira nomes significativos.
- Evite modificar a lista enquanto a percorre.
- Utilize compreensão de listas quando simplificar o código.
- Use métodos apropriados em vez de reinventar funcionalidades.


```python
compras = ["arroz", "feijão", "leite"]

if "leite" in compras:
    print("Comprar leite")
```

# Exercícios

1. Crie uma lista com 10 números e mostre o maior e o menor.

```python
lista = [1,2,3,4,5]
print(f'{max(lista)} é o maior número da lista')
print(f'{min(lista)} é o menor número da lista')
```

2. Leia 5 nomes e armazene em uma lista.

```python
nome1 = str(input("Digite o primeiro nome:"))
nome2 = str(input("Digite o segundo nome:"))
nome3 = str(input("Digite o terceiro nome:"))
nome4 = str(input("Digite o quarto nome:"))
nome5 = str(input("Digite o quinto nome:"))


lista = [nome1, nome2, nome3, nome4, nome5]
print(lista)
```

3. Conte quantas vezes um número aparece na lista.

```python
lista = [1,2,3,4,5,5]
len([x for x in lista if lista.count(x) > 1])
```

4. Ordene uma lista em ordem crescente e decrescente.

```python
lista = [1, 2, 3, 4, 5, 324, 23423, 3242354]
lista.sort()
print(lista)
lista.reverse()
print(lista)
```
