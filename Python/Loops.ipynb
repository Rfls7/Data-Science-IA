# Loops em Python

Este notebook explica em detalhes os **laços de repetição (loops)** em Python, incluindo `for`, `while`, controle de fluxo e boas práticas.

## 1. O que são loops?

Loops permitem executar um bloco de código várias vezes, evitando repetição de código e facilitando a automação de tarefas.

Os principais laços em Python são:
- `for`
- `while`


```python
for i in range(3):
    print("Olá!")
```

## 2. O laço `for`

O `for` percorre sequências como listas, strings, tuplas, dicionários e objetos `range()`.

```python
nomes = ["Ana", "Carlos", "Maria"]

for nome in nomes:
    print(nome)
```

## 3. A função `range()`

`range()` gera uma sequência de números.

Sintaxes:
- `range(fim)`
- `range(inicio, fim)`
- `range(inicio, fim, passo)`


```python
for i in range(10, 0, -2):
    print(i)
```

## 4. O laço `while`

O `while` executa enquanto a condição for verdadeira.

```python
contador = 1

while contador <= 5:
    print(contador)
    contador += 1
```

## 5. `break`

Encerra imediatamente o loop.

```python
for numero in range(10):
    if numero == 5:
        break
    print(numero)
```

## 6. `continue`

Pula a iteração atual e continua para a próxima.

```python
for numero in range(6):
    if numero == 3:
        continue
    print(numero)
```

## 7. `pass`

`pass` funciona como um espaço reservado quando um bloco ainda não possui implementação.

```python
for i in range(3):
    pass

print("Loop finalizado")
```

## 8. Loops aninhados

É possível colocar um loop dentro de outro.

```python
for linha in range(3):
    for coluna in range(4):
        print(f"({linha},{coluna})", end=" ")
    print()
```

## 9. `else` em loops

O bloco `else` é executado quando o loop termina normalmente (sem `break`).

```python
for i in range(3):
    print(i)
else:
    print("Loop concluído")
```

## 11. Boas práticas

- Evite loops infinitos.
- Atualize corretamente a variável de controle no `while`.
- Utilize `for` quando souber a quantidade de repetições.
- Utilize `while` quando depender de uma condição.
- Prefira código simples e legível.


```python
senha = ""

while senha != "python":
    senha = input("Digite a senha: ")

print("Acesso permitido!")
```
