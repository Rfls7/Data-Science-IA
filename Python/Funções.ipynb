Este notebook apresenta os principais conceitos sobre **funções em Python**, com teoria, exemplos e exercícios.

## 1. O que é uma função?

Uma função é um bloco de código reutilizável que executa uma tarefa específica. Ela evita repetição de código e torna os programas mais organizados.

```python
#Definindo uma função em Python
def contagem():
    result = 1 + 1 
    return result

#Chamando a função
contagem()
```

## 2. Criando funções

A palavra-chave `def` define uma função.

```python
def somar(a, b):
    return a + b

resultado = somar(5, 7)
print(resultado)
```

## 3. Parâmetros e argumentos

```python
def apresentar(nome, idade):
    print(f"Nome: {nome}")
    print(f"Idade: {idade}")

apresentar("Maria", 30)
```

## 4. Valores padrão

```python
def potencia(base, expoente=2):
    return base ** expoente

print(potencia(5))
print(potencia(5, 3))
```

## 5. Argumentos nomeados

```python
def cadastro(nome, cidade):
    print(nome, cidade)

cadastro(cidade="Brasília", nome="Ana")
```

## 6. Retornando valores

```python
def media(n1, n2):
    return (n1+n2)/2

m = media(8, 10)
print(m)
```

## 7. Escopo de variáveis

```python
# Uma variável definida fora da função é global e pode ser acessada em qualquer parte do código
x = 10

# Uma variável definida dentro da função é local e não pode ser acessada fora dela
def exemplo():
    x = 5
    print("Local:", x)

# Na função exemplo, a variável x é local e não afeta a variável global x
exemplo()

# No print abaixo, a variável x é global e tem o valor 10
print("Global:", x)
```

## 8. Funções com múltiplos retornos

```python
def operacoes(a,b):
    return a+b, a-b, a*b

soma, sub, mult = operacoes(10,5)
print(soma, sub, mult)
```

## 9. *args

```python
def soma_total(*numeros):
    return sum(numeros)

print(soma_total(1,2,3,4,5))
```

## 10. **kwargs

```python
def perfil(**dados):
    for chave, valor in dados.items():
        print(chave, "=", valor)

perfil(nome="Carlos", idade=40, profissao="Professor")
```

## 11. Funções lambda

Quando usar lambda?

Funções lambda são indicadas para funções simples e curtas, normalmente utilizadas como argumento de outras funções, como map(), filter() e sorted(). Quando a lógica é mais complexa, é preferível utilizar def, pois o código fica mais legível e fácil de manter.

```python
# Cria uma função anônima (lambda) que recebe um parâmetro chamado 'x'
# e retorna o valor de x elevado ao quadrado (x²).
quadrado = lambda x: x**2

# Chama a função passando o valor 8 como argumento
# e imprime o resultado na tela.
print(quadrado(8))
```

## 12. Documentação (docstring)

```python
def area_retangulo(base, altura):
    """Calcula a área de um retângulo."""
    return base * altura


# Chama a função help() para exibir a documentação da função area_retangulo.
help(area_retangulo)

# Chama a função sem mostrar o valor comentado, apenas para calcular a área do retângulo com base 2 e altura 3.
area_retangulo(2, 3)
```

## 13. Boas práticas

- Use nomes descritivos.
- Faça funções pequenas.
- Evite efeitos colaterais.
- Documente funções importantes.
- Prefira retornar valores em vez de imprimir quando possível.

## 14. Resumo dos conceitos

| Conceito | Exemplo |

| Definir função | `def nome():` |

## 15. Exercícios

1. Crie uma função que calcule o dobro de um número.

```python
def dobro(x):
    """Calcula o dobro de um número."""
    return 2 * x

dobro(5)
```

2. Crie uma função que verifique se um número é par.

```python
def par_ou_impar(numero):
    """Verifica se um número é par ou ímpar."""
    if numero % 2 == 0:
        return "Par"
    else:
        return "Ímpar"
```

3. Faça uma função que receba uma lista e retorne o maior valor.

```python
def lista(elementos):  
    return max(elementos)

lista([1, 2, 3, 4, 5])
```
