## Limpeza de Dados

### Exercício 21

### Exercício 22
**Enunciado:** Preencha valores nulos com a média.

````python
# Sua solução aqui

# Seleciona apenas colunas numéricas
num_cols = df.select_dtypes(include=["int64", "float64"]).columns

# Preenche os nulos com a média de cada coluna
for col in num_cols:
    df[col] = df[col].fillna(df[col].mean())
````

### Exercício 23
**Enunciado:** Preencha valores categóricos nulos.

````python
# Sua solução aqui

# Seleciona apenas colunas categóricas
cat_cols = df.select_dtypes(include=["object", "category"]).columns

# Preenche os nulos com a moda de cada coluna
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
````

### Exercício 30
**Enunciado:** Identifique outliers usando IQR.

````python
# Seleciona colunas numéricas
num_cols = df.select_dtypes(include='number').columns

# Máscara para identificar linhas com outliers
mask_outliers = pd.Series(False, index=df.index)

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    mask_outliers |= (
        (df[col] < limite_inferior) |
        (df[col] > limite_superior)
    )

# DataFrame contendo apenas os outliers
outliers = df[mask_outliers]

print(outliers)
````
