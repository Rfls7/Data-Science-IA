# O que é uma tupla?

Uma tupla é uma coleção de elementos:

Ordenada
Imutável (não pode ser alterada após criada)
Permite elementos repetidos
Pode armazenar diferentes tipos de dados

Exemplo:

```python
cores = ("azul", "verde", "vermelho")
print(cores)
```

## Criando tuplas
### Tupla comum

```python
frutas = ("maçã", "banana", "uva")
print(frutas)
```

### Tupla com diferentes tipos

```python
dados = ("João", 25, 1.75, True)

print(dados)
```

### Tupla vazia

```python
vazia = ()

print(vazia)
```

### Tupla com apenas um elemento

⚠️ Atenção!

Sem a vírgula, o Python entende como outro tipo de dado.

```python
numero = (10,)
print(type(numero))
```

```python
numero = (10)
print(type(numero))
```

## Acessando elementos

```python
animais = ("gato", "cachorro", "coelho")

print(animais[0])
print(animais[1])
print(animais[2])
```

### Índices negativos

```python
print(animais[-1])
print(animais[-2])
```

### Fatiamento (Slicing)

```python
numeros = (10, 20, 30, 40, 50)

print(numeros[1:4])
```

```python
print(numeros[:3])
```

```python
print(numeros[2:])
```

### index()
* Acessa o elemento da tupla, lembrando que a contagem começa em 0

```python
print(numeros.index(3))
```

### len()
* Conta quantos arquivos há dentro da tupla

```python
print(len(numeros))
```

### max()
* Calcula o valor máximo da tupla

```python
print(max(numeros))
```

### min()
* Calcula o valor mínimo da tupla

```python
print(min(numeros))
```

### sum()
* Calcula a soma dos valores da tupla

```python
print(sum(numeros))
```

### Desempacotamento (Unpacking)

```python
dados = ("Maria", 30, "Analista")

nome, idade, profissao = dados

print(nome)
print(idade)
print(profissao)
```

#### Ignorando valores

```python
dados = ("Ana", 28, "Engenheira")

nome, _, profissao = dados

print(nome)
print(profissao)
```
